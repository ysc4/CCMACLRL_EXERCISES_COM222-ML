## Exercise 2

Machine Learning pipeline

1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

2. Load the dataset


```python
#put your answer here

dataframe = pd.read_csv("iris.csv")

dataframe.drop(columns='Id', inplace=True)
dataframe.head()

dataframe.info()

dataframe['Species'].value_counts()
```

3. Visualize the dataset

```python
# put your answer here

plt.hist(dataframe["SepalLengthCm"])
plt.xlabel("SepalLength")
plt.ylabel("Number of rows")
plt.show()
```

```python
plt.hist(dataframe["PetalLengthCm"])
plt.xlabel("PetalLength")
plt.ylabel("Number of rows")
plt.show()
```

```python
fig, ax = plt.subplots(2,2, figsize=(8,6))

ax[0,0].hist(dataframe["SepalLengthCm"])
ax[0,1].hist(dataframe["SepalWidthCm"])
ax[1,0].hist(dataframe["PetalLengthCm"])
ax[1,1].hist(dataframe["PetalWidthCm"])

ax[0,0].set_title("sepal length")
ax[0,1].set_title("sepal width")
ax[1,0].set_title("petal length")
ax[1,1].set_title("petal width")

plt.show()
```

```python
colors = {'Iris-setosa': 'red','Iris-virginica': 'blue', 'Iris-versicolor': 'green'}
```

```python
plt.scatter(dataframe["SepalLengthCm"], dataframe["SepalWidthCm"], c=dataframe['Species'].map(colors))

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

```python
plt.scatter(dataframe["PetalLengthCm"], dataframe["PetalWidthCm"], c=dataframe['Species'].map(colors))

plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.show()
```

```python
plt.scatter(dataframe["SepalLengthCm"], dataframe["PetalLengthCm"], c=dataframe['Species'].map(colors))

plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.title("Iris Data Scatter Plot")

plt.show()
```

4. Split the dataset into a training set and a testing set

```python
# put your answer here

X = dataframe.drop(columns=['Species']).values
y = dataframe['Species'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

5. Train the model

```python
# put your answer here

knn = KNeighborsClassifier(n_neighbors=13)

knn.fit(X_train, y_train)
```

6. Test the model

```python
# put your answer here

y_pred = knn.predict(X_test)
metrics.accuracy_score(y_test, y_pred)
```
